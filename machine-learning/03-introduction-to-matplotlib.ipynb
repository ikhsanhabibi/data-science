## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11,22,33,44]

# 2. Setup plot
fix, ax = plt.subplots(figsize=(5,5))

# 3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show figure
fig = plt.gcf() # get current figure
fig.savefig("./images/sample-plot.png")
plt.show()
```

## two options for subplots
## conclusion : we can have multiple plots as one figure

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows =2, ncols=2, figsize=(10, 5))

#plot each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
# subplots opion 2
fig, ax = plt.subplots(nrows =2, ncols=2, figsize=(10, 5))

# plot each diffrent index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
```

### Which graph should be used? pyplot vs matplotlib OO method

* when plotting smtg quickly => pyplot
* when plotting more advanced => matplotlib object oriented

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.size
```

```python
over_50.head()
```

```python
# rows length 
len(heart_disease)
```

```python
len(over_50)
```

```python
# Pyplot method (it is not good)
over_50.plot(kind="scatter", x="age", y="chol", c="target", figsize=(10,6))
```

```python
# OO matplotlib method
fix, ax = plt.subplots(figsize= (10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)

# limit the range x axis
#ax.set_xlim([45, 100]);
```

```python
#OO matplotlib from the scratch
fix, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"], c=over_50["target"])

# customize the plot
ax.set(title="Heart disease and cholesterol levels", 
      xlabel="Age", ylabel="Cholesterol")

# add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# add a horizontal line
ax.axhline(y=over_50["chol"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalac

# single figure, 2 axis
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10),
                              sharex=True)

# add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"],
                      c=over_50["target"])
# custom ax0
ax0.set(title="Heart disease and cholesterol levels",
       ylabel="Cholesterol")
# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# add meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");


# add data to ax1
scatter = ax1.scatter(x=over_50["age"],y=over_50["thalach"],
                      c=over_50["target"] )
# custom ax1
ax1.set(title="Heart disease and max heart rate",
       xlabel="Age", ylabel="Max Heart Rate");
# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# add meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");


# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

## Using another style

```python
plt.style.use("seaborn-whitegrid")

#OO matplotlib from the scratch
fix, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"], c=over_50["target"],
                    cmap="winter") # this changes the color scheme

# customize the plot
ax.set(title="Heart disease and cholesterol levels", 
      xlabel="Age", ylabel="Cholesterol")

# add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# add a horizontal line
ax.axhline(y=over_50["chol"].mean(), linestyle="--");
```

```python
# customize the sytle

# Subplot of chol, age, thalac

# single figure, 2 axis
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10),
                              sharex=True)

# add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"],
                      c=over_50["target"], cmap="winter")
# custom ax0
ax0.set(title="Heart disease and cholesterol levels",
       ylabel="Cholesterol")
ax0.set_xlim(50, 80)

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# add meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");


# add data to ax1
scatter = ax1.scatter(x=over_50["age"],y=over_50["thalach"],
                      c=over_50["target"], cmap="winter" )
# custom ax1
ax1.set(title="Heart disease and max heart rate",
       xlabel="Age", ylabel="Max Heart Rate");
ax1.set_xlim(50, 80)
ax1.set_ylim(60, 200)

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# add meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");


# add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

## Save the figure image
plt.savefig('heart-disease-analysis-plot.png')
```
